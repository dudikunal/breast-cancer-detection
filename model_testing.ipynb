# Breast cancer Detection
In this Notebook we try to optmize models for breast cancer Classification

Dataset: https://www.kaggle.com/datasets/uciml/breast-cancer-wisconsin-data?resource=download

## Importing Libraries

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
```

## Preprocessing

### Importing dataset

```python
data_path = 'dataset/data.csv'
df = pd.read_csv(data_path)
df.head()
```

```python
col1 = df.iloc[ : , 1 : 5]
col1
```

### Encoding Categorical data

```python
# Removing the first and last col
df = df.iloc[:, 1:-1]
```

```python
diagnosis_col = df['diagnosis']
lb = LabelEncoder()
lb.fit(diagnosis_col)
diagnosis_col = lb.transform(diagnosis_col)
df['diagnosis'] = diagnosis_col
df
```

### Test Train Split

```python
X = df.iloc[:, 1:] # Features
y = df['diagnosis'] # Target
```

```python
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
print(X_train.shape, X_test.shape, " " , y_train.shape, y_test.shape)
```

### Gaussian Naive Bayes 
Now, We **Fit** the Model on **Training** Data and **Evaluate** the Model on **Testing** Data

```python
# Making object of Gaussian Navive Bayes Classifier
gnb = GaussianNB()

# Training / Fitting
gnb.fit(X_train, y_train)
```

```python
y_pred = gnb.predict(X_test)
y_pred
```

```python
results = {'y' : y_test, 'y_pred' : y_pred}
results = pd.DataFrame(results)
print("Number of mislabeled points out of a total %d points : %d"
      % (X_test.shape[0], (y_test != y_pred).sum()))
results
```

```python
from sklearn.metrics import classification_report

# report = pd.DataFrame()
report = classification_report(y_test, y_pred, target_names=['Diseased', 'Normal'], output_dict=True)
report = pd.DataFrame(report)
report
```
